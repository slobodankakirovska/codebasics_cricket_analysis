### Primary Insights


### 1. Top 10 batsmen based on past 3 years total runs scored.


```python
# Convert 'matchDate' to datetime
merged_data['matchDate'] = pd.to_datetime(merged_data['matchDate'])

# Filter data for past 3 years
three_years_ago = pd.to_datetime('today') - pd.DateOffset(years=3)
merged_data = merged_data[merged_data['matchDate'] >= three_years_ago]


# Task 1: Top 10 batsmen based on past 3 years total runs scored
top_batsmen_total_runs = merged_data.groupby('batsmanName')['runs_x'].sum().nlargest(10)
print("Task 1: Top 10 batsmen based on past 3 years total runs scored:")
print(top_batsmen_total_runs)
top_batsmen_total_runs.to_csv('top_batsmen_total_runs.csv')
```

### 2. Top 10 batsmen based on past 3 years batting average. (min 60 balls faced in each season)


```python
# Assuming your DataFrame is named merged_data
# Group by batsmanName and season, and calculate total runs and total times dismissed
season_stats = merged_data.groupby(['batsmanName', merged_data['matchDate'].dt.year]).agg({'runs_x': 'sum', 'out/not_out': 'count'})

# Filter out batsmen who have faced at least 60 balls in each season
qualified_batsmen = season_stats.groupby('batsmanName').filter(lambda x: (x['out/not_out'] >= 60).all())

# Calculate the number of times not out for each batsman
not_outs = merged_data[merged_data['out/not_out'] == 'not_out'].groupby('batsmanName')['out/not_out'].count()

# Merge not_outs into qualified_batsmen
qualified_batsmen = pd.merge(qualified_batsmen, not_outs, left_index=True, right_index=True, how='left')
qualified_batsmen.rename(columns={'out/not_out_x': 'dismissals', 'out/not_out_y': 'not_outs'}, inplace=True)

# Adjust total dismissals by adding the number of times not out
qualified_batsmen['total_dismissals'] = qualified_batsmen['dismissals'] + qualified_batsmen['not_outs']

# Calculate batting average for each batsman
qualified_batsmen['batting_average'] = qualified_batsmen['runs_x'] / qualified_batsmen['total_dismissals']

# Sort batsmen based on batting average
top_batsmen_batting_average = qualified_batsmen.groupby('batsmanName')['batting_average'].mean().sort_values(ascending=False).head(10)

print("Task 2: Top 10 batsmen based on batting average in the past 3 years (minimum 60 balls faced in each season):")
print(top_batsmen_batting_average)
top_batsmen_batting_average.to_csv('top_batsmen_batting_average.csv')
```

### 3. Top 10 batsmen based on past 3 years strike rate (min 60 balls faced in each season)


```python
# Assuming your DataFrame is named merged_data
# Convert matchDate to datetime type
merged_data['matchDate'] = pd.to_datetime(merged_data['matchDate'])

# Filter data for the past 3 years
three_years_ago = pd.to_datetime('today') - pd.DateOffset(years=3)
merged_data_recent = merged_data[merged_data['matchDate'] >= three_years_ago]

# Group by batsmanName and season, and calculate total runs, balls faced, wides, and no-balls
season_stats = merged_data_recent.groupby(['batsmanName', merged_data_recent['matchDate'].dt.year]).agg({'runs_x': 'sum', 'balls': 'sum', 'wides': 'sum', 'noBalls': 'sum'})

# Filter out batsmen who have faced at least 60 balls in each season
qualified_batsmen = season_stats.groupby('batsmanName').filter(lambda x: (x['balls'] >= 60).all())

# Calculate legitimate balls faced
qualified_batsmen['legitimate_balls'] = qualified_batsmen['balls'] - (qualified_batsmen['wides'] + qualified_batsmen['noBalls'])

# Filter out batsmen who have faced less than 60 legitimate balls in any season
qualified_batsmen = qualified_batsmen.groupby('batsmanName').filter(lambda x: (x['legitimate_balls'] >= 60).all())

# Calculate strike rate for each batsman
qualified_batsmen['strike_rate'] = (qualified_batsmen['runs_x'] / qualified_batsmen['legitimate_balls']) * 100

# Sort batsmen based on strike rate
top_batsmen_strike_rate = qualified_batsmen.groupby('batsmanName')['strike_rate'].mean().sort_values(ascending=False).head(10)

print("Task 3: Top 10 batsmen based on strike rate in the past 3 years (minimum 60 balls faced in each season):")
print(top_batsmen_strike_rate)
top_batsmen_strike_rate.to_csv('top_batsmen_strike_rate.csv')
```

### 4. Top 10 bowlers based on past 3 years total wickets taken.

```python
# Assuming your DataFrame is named merged_data
# Convert matchDate to datetime type
merged_data['matchDate'] = pd.to_datetime(merged_data['matchDate'])

# Filter data for the past 3 years
three_years_ago = pd.to_datetime('today') - pd.DateOffset(years=3)
merged_data_recent = merged_data[merged_data['matchDate'] >= three_years_ago]

# Group by bowlerName and calculate total wickets taken
bowler_stats = merged_data_recent.groupby('bowlerName')['wickets'].sum()

# Sort bowlers based on total wickets taken
top_bowlers_wickets = bowler_stats.sort_values(ascending=False).head(10)

print("Task 4: Top 10 bowlers based on total wickets taken in the past 3 years:")
print(top_bowlers_wickets)
top_bowlers_wickets.to_csv('top_bowlers_wickets.csv')
```

### 5. Top 10 bowlers based on past 3 years bowling average. (min 60 balls bowled in each season)

```python
import pandas as pd

# Assuming your DataFrame is named merged_data
# Convert matchDate to datetime type
merged_data['matchDate'] = pd.to_datetime(merged_data['matchDate'])

# Filter data for the past 3 years
three_years_ago = pd.to_datetime('today') - pd.DateOffset(years=3)
merged_data_recent = merged_data[merged_data['matchDate'] >= three_years_ago]

# Group by bowlerName and season, and calculate total runs conceded and total wickets taken
season_stats = merged_data_recent.groupby(['bowlerName', merged_data_recent['matchDate'].dt.year]).agg({'runs_y': 'sum', 'wickets': 'sum', 'balls': 'sum'})

# Filter out bowlers who have bowled at least 60 balls in each season
qualified_bowlers = season_stats.groupby('bowlerName').filter(lambda x: (x['balls'] >= 60).all())

# Calculate bowling average for each bowler
qualified_bowlers['bowling_average'] = qualified_bowlers['runs_y'] / qualified_bowlers['wickets']

# Sort bowlers based on bowling average
top_bowlers_bowling_average = qualified_bowlers.groupby('bowlerName')['bowling_average'].mean().sort_values().head(10)

print("Task 5: Top 10 bowlers based on bowling average in the past 3 years (minimum 60 balls bowled in each season):")
print(top_bowlers_bowling_average)
top_bowlers_bowling_average.to_csv('top_bowlers_bowling_average.csv')
```

### 6. Top 10 bowlers based on past 3 years economy rate. (min 60 balls bowled in each season)


```python
import pandas as pd

# Assuming your DataFrame is named merged_data
# Convert matchDate to datetime type
merged_data['matchDate'] = pd.to_datetime(merged_data['matchDate'])

# Filter data for the past 3 years
three_years_ago = pd.to_datetime('today') - pd.DateOffset(years=3)
merged_data_recent = merged_data[merged_data['matchDate'] >= three_years_ago]

# Group by bowlerName and season, and calculate total runs conceded and total balls bowled
season_stats = merged_data_recent.groupby(['bowlerName', merged_data_recent['matchDate'].dt.year]).agg({'runs_y': 'sum', 'balls': 'sum'})

# Filter out bowlers who have bowled at least 60 balls in each season
qualified_bowlers = season_stats.groupby('bowlerName').filter(lambda x: (x['balls'] >= 60).all())

# Calculate economy rate for each bowler
qualified_bowlers['economy_rate'] = qualified_bowlers['runs_y'] / (qualified_bowlers['balls'] / 6)  # 6 balls per over

# Sort bowlers based on economy rate
top_bowlers_economy_rate = qualified_bowlers.groupby('bowlerName')['economy_rate'].mean().sort_values().head(10)

print("Top 10 bowlers based on economy rate in the past 3 years (minimum 60 balls bowled in each season):")
print(top_bowlers_economy_rate)

top_bowlers_economy_rate.to_csv('top_bowlers_economy_rate.csv')
```

### 7. Top 5 batsmen based on past 3 years boundary % (fours and sixes).


```python
# Assuming your DataFrame is named merged_data
# Convert matchDate to datetime type
merged_data['matchDate'] = pd.to_datetime(merged_data['matchDate'])

# Filter data for the past 3 years
three_years_ago = pd.to_datetime('today') - pd.DateOffset(years=3)
merged_data_recent = merged_data[merged_data['matchDate'] >= three_years_ago]

# Group by batsmanName and calculate total fours and sixes scored and total balls faced
batsman_stats = merged_data_recent.groupby('batsmanName').agg({'4s_x': 'sum', '6s_x': 'sum', 'balls': 'sum'})

# Calculate total runs scored by each batsman
batsman_stats['runs'] = (batsman_stats['4s_x'] * 4) + (batsman_stats['6s_x'] * 6)

# Calculate boundary percentage for each batsman
batsman_stats['boundary_percentage'] = ((batsman_stats['4s_x'] + batsman_stats['6s_x']) / batsman_stats['runs']) * 100

# Sort batsmen based on boundary percentage
top_batsmen_fours_sixes = batsman_stats['boundary_percentage'].nlargest(5)

print("Top 5 batsmen based on boundary percentage in the past 3 years (fours and sixes):")
print(top_batsmen_fours_sixes)
top_batsmen_fours_sixes.to_csv('top_batsmen_fours_sixes.csv')
```

### 8. Top 5 bowlers based on past 3 years dot ball %.


```python
# Assuming your DataFrame is named merged_data
# Convert matchDate to datetime type
merged_data['matchDate'] = pd.to_datetime(merged_data['matchDate'])

# Filter data for the past 3 years
three_years_ago = pd.to_datetime('today') - pd.DateOffset(years=3)
merged_data_recent = merged_data[merged_data['matchDate'] >= three_years_ago]

# Group by bowlerName and calculate total dot balls bowled and total balls bowled
bowler_stats = merged_data_recent.groupby('bowlerName').agg({'0s': 'sum', 'balls': 'sum'})

# Calculate dot ball percentage for each bowler
bowler_stats['dot_ball_percentage'] = (bowler_stats['0s'] / bowler_stats['balls']) * 100

# Sort bowlers based on dot ball percentage
top_bowlers_dot_ball = bowler_stats['dot_ball_percentage'].nlargest(5)

print("Top 5 bowlers based on dot ball percentage in the past 3 years:")
print(top_bowlers_dot_ball)
top_bowlers_dot_ball.to_csv('top_bowlers_dot_ball.csv')
```

### 9. Top 4 teams based on past 3 years winning %.

```python
# Assuming your DataFrame is named merged_data
# Convert matchDate to datetime type
merged_data['matchDate'] = pd.to_datetime(merged_data['matchDate'])

# Filter data for the past 3 years
three_years_ago = pd.to_datetime('today') - pd.DateOffset(years=3)
merged_data_recent = merged_data[merged_data['matchDate'] >= three_years_ago]

# Group by team1 and calculate total matches played and total matches won
team1_stats = merged_data_recent.groupby('team1').agg({'match_id': 'nunique', 'winner': lambda x: (x == 'team1').sum()})
team1_stats.rename(columns={'match_id': 'matches_played', 'winner': 'matches_won'}, inplace=True)

# Group by team2 and calculate total matches played and total matches won
team2_stats = merged_data_recent.groupby('team2').agg({'match_id': 'nunique', 'winner': lambda x: (x == 'team2').sum()})
team2_stats.rename(columns={'match_id': 'matches_played', 'winner': 'matches_won'}, inplace=True)

# Merge the results from both groups
team_stats = team1_stats.add(team2_stats, fill_value=0)

# Calculate winning percentage for each team
team_stats['winning_percentage'] = (team_stats['matches_won'] / team_stats['matches_played']) * 100

# Sort teams based on winning percentage
top_teams = team_stats['winning_percentage'].nlargest(4)

print("Top 4 teams based on winning percentage in the past 3 years:")
print(top_teams)
top_teams.to_csv('top_teams.csv')
```
